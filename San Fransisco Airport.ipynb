```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

keterangan:

01L/R = Jumlah keberangkatan dari landasan pacu 01L/R. 

10L/R = Jumlah keberangkatan dari landasan pacu 10L/R. 

19L/R = Jumlah keberangkatan dari landasan pacu 19L/R. 

28L/R = Jumlah keberangkatan dari landasan pacu 28L/R. 

```python
df = pd.read_csv("late-night-preferential-runway-use-1.csv")
df = df.rename

df
```

```python
df_group = df.groupby(['Month'])[['01L/R', '10L/R', '19L/R', '28L/R']].sum()
df_baru = df_group.rename(index={1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 7:"July", 8:"August", 9:"September", 10:"Oktober", 11:"November", 12:"Desember"})
df_baru
```

visualisasi data yang telah dikelompokkan berdasarkan bulan menggunakan line plot

```python
df_baru.plot(kind='line', figsize=(14,6), marker='o')
plt.title('Jumlah Keberangkatan San Francisco Airport')
plt.xlabel('\nMonth')
plt.ylabel('Jumlah Keberangkatan')
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(prop={'size': 12})
plt.show()
```

Dari visualisasi data jumlah keberangkatan  pesawat San Fransisco Airport

1. Jumlah keberangkatan terbanyak terjadi pada bulan ke-6 atau bulan Juni pada landasan Pacu 28L/R

2. landasan pacu yang mengalami perubahan menurun paling tinggi adalah landasan pacu 28L/R pada bulan ke-6 sampai bulan ke-12




```python

df_total12 = df_baru.drop(columns=['10L/R', '19L/R', '28L/R'])
df_total123 = pd.DataFrame(df_total12).reset_index()
df_total123
df_total123
```

```python
plt.figure(figsize=(18,6))
plt.bar(df_total123['Month'], df_total123['01L/R'], color='aqua')
plt.title('Jumlah Keberangkatan San Francisco Airport landasan pacu 01L/R')
plt.xlabel('\nBulan', size=15)
plt.ylabel('Jumlah\n', size=15)
plt.xticks(size=14)
plt.yticks(size=14)
plt.show()
```
